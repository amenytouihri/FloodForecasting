# Model: ARIMA 

```python
# imports
from pmdarima import auto_arima
import pandas as pd
```

```python
data = pd.read_csv('../dataset/barrages.csv')
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.Date.apply(lambda x: x.strftime('%Y%m%d')).astype(int)
df = data.set_index('Date')
y = df['stock']
```

```python
arima_model = auto_arima(y, seasonal=True, m=7)
y_arima_forecast = arima_model.predict(n_periods=365)
```
