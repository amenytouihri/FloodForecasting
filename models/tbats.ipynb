# Model: ARIMA 

```python
# imports
import pandas as pd
```

```python
data = pd.read_csv('../dataset/cleanedbarrage.csv')
data
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.Date.apply(lambda x: x.strftime('%Y%m%d')).astype(int)
df = data.set_index('Date')
y = df['stock']
```

The first thing is to identify if the data is stationary or non – stationary.

To identify the nature of data, we will be using the null hypothesis.

H0: The null hypothesis: It is a statement about the population that either is believed to be true or is used to put forth an argument unless it can be shown to be incorrect beyond a reasonable doubt.

H1: The alternative hypothesis: It is a claim about the population that is contradictory to H0 and what we conclude when we reject H0.

#Ho: It is non-stationary
#H1: It is stationary


```python
y_to_train = y.iloc[:(len(y)-365)]
y_to_test = y.iloc[(len(y)-365):]
```

```python
from tbats import TBATS, BATS
# Fit the model
estimator = TBATS(seasonal_periods=(7, 365.25))
model = estimator.fit(y_to_train)
# Forecast 365 days ahead
y_forecast = model.forecast(steps=365)
```

```python
# Create estimator
estimator = TBATS(
    seasonal_periods=[14, 30.5],
    use_arma_errors=False,  # shall try only models without ARMA
    use_box_cox=False  # will not use Box-Cox
)
model = estimator.fit(y)
```

```python
# Short model summary
print('\n\nSUMMARY FUNCTION\n\n')
print(model.summary())
```

```python
print('\n\nIN SAMPLE PREDICTIONS\n\n')
print('Original time series (5 first values)', model.y[:5])
print('Predictions (5 first values)', model.y_hat[:5])
print('Residuals (5 first values)', model.resid[:5])
```

```python
import numpy as np
print('\n\nFORECAST\n\n')
print('Values', y_to_test)
print('MAE', np.mean(np.abs(y_forecast - y_to_test)))
```
